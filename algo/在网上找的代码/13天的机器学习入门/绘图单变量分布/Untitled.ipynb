# 绘制双变量分布

## 2.1 绘制散点图

```python
# 创建DataFrame对象
df = pd.DataFrame({"X": np.random.randn(500), "y": np.random.randn(500)})
# 绘制散布图
sns.jointplot(df,kind="scatter",color="r",size=10,ratio=5,space=1)
```

# 绘制二维直方图

```python
# 创建DataFrame对象
df = pd.DataFrame({"X": np.random.randn(500), "y": np.random.randn(500)})
# 绘制散布图
sns.jointplot(df,kind="kde",color="r",size=10,ratio=5,space=1)
```
